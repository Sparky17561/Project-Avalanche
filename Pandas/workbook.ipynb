```python
# concatenate and merge dataframe (JOINS)

df1 = pd.DataFrame({'ID':[1,2,3],'Name':['A','B','C']})
df1
```

```python
# concatenate and merge dataframe (JOINS)

df2 = pd.DataFrame({'ID':[1,2,2,4],'Score':[88,96,77,79]})
df2
```

```python
pd.concat([df1,df2],axis=0) # ek k neeche ek row
```

```python
pd.concat([df1,df2],axis=1) # ek k neeche ek col
```

```python
pd.merge(df1,df2,how='inner',on='ID')
```

```python
pd.merge(df1,df2,how='inner',left_on='ID',right_on='ID')
```
